```python
import pandas as pd 
```

```python
#Dataset details: https://archive.ics.uci.edu/ml/datasets/Epileptic+Seizure+Recognition
```

```python
d = pd.read_csv('initialData.csv')
d
```

```python
#y=1 means patient is having seizure. Any other number means that they're not having a seizure (see the dataset description for more details)
#So to make this a binary problem I'll turn every non-1 value into a 0 

d['class'] = d.y==1  #note that d.y==1 is a elementwise boolean comparison 
d['class'] = d['class'].astype(int) #converts the booleans made in the line above to ints 

#Removing the original y and the useless patient IDs. 
#Note that .pop('y') is the same as .drop('y', axis=1,inplace=True) but is more concise :)
d.pop('y')
d.pop('Unnamed: 0')
print()
```

```python
#Quick test with an SVM: 
```

```python
y = d['class'] #sets y to be class column 
X = d.iloc[:,0:(d.shape[1]-1)] #sets X to be dataset with class column removed 
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 
```

```python
from sklearn import svm
modelSVM = svm.SVC(gamma=0.001, C=100.) 
modelSVM.fit(X_train, y_train)
print(classification_report(y_test,modelSVM.predict(X_test)))
```

```python
#80% is pretty bad considering the fact that we only have two classes 
#and one of them makes up 80% of the dataset...
```

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=1)
modelRF.fit(X_train, y_train)
print(classification_report(y_test,modelRF.predict(X_test)))
```

```python
#97% overall accuracy and 92% seizure prediction
#possibly good baseline to test against
```

```python
from sklearn.neural_network import MLPClassifier
modelMLP = MLPClassifier(solver='adam', alpha=1e-5, learning_rate='adaptive', hidden_layer_sizes=(5,2), random_state=2)
modelMLP.fit(X_train, y_train)
print(classification_report(y_test,modelMLP.predict(X_test)))
```

```python
#95% overall accuracy and a 93% seizure prediction
#setting the solver to adam and learning rate as adaptive improve accuracy from the low 80s
#neural networks and random forest seem promising out of the 3 baselines implemented
#we can potentially look at more advanced models and tuning
```
